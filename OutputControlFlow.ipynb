```python
import numpy as np
import pygame
import time
import pygame.freetype

model_accuracy_as = 0.8
GREEN = (0, 255, 0)
RED = ( 255, 0, 0)
```

```python
fur_elise = []
for i in range(1,3):
    fur_elise.append("FurElise" + str(i) +".wav")
    
def sound_play(file):
    print("   Playing " + str(file))
    #load in and initialize the music file 
    pygame.init()
    pygame.mixer.init()
    pygame.mixer.music.load(file)

    #This would be the part you do at the end of each block 
    pygame.mixer.music.play()
    while pygame.mixer.music.get_busy():
        pygame.time.Clock().tick(10)

def play_all(crec,colrec, string):
    for i in range(0,len(crec)):
        display(string, colrec[i])
        sound_play(str(crec[i]))
        
def play(x):
    sound_play(str(x))

def task_binary_classifier():
    return np.random.binomial(n=1, p=model_accuracy_as, size=1)
    
def display(string,col = (255,255,255)):
    pygame.init()
    screen = pygame.display.set_mode((1300, 700))
    GAME_FONT = pygame.freetype.Font("/usr/share/fonts/truetype/freefont/FreeMono.ttf", 24)
    running =  True
    screen.fill(col)
    GAME_FONT.render_to(screen, (650 - 8 * len(string),350), string, (0,0,0))
    pygame.display.flip()
```
